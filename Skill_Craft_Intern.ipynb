```python
!pip install pandas matplotlib seaborn

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Country': ['India', 'USA', 'Norway', 'Brazil', 'Nigeria', 'Japan', 'Germany', 'Egypt'],
    'Female_Labor_Participation_Rate': [20.3, 56.7, 61.5, 52.4, 48.1, 51.3, 54.6, 23.5],
    'GDP_per_Capita': [2100, 65000, 75000, 8900, 2400, 40000, 47000, 3800],
    'Literacy_Rate_Female': [65, 99, 99, 92, 58, 99, 99, 70],
    'Fertility_Rate': [2.2, 1.7, 1.5, 1.8, 5.3, 1.4, 1.6, 3.4],
    'Urban_Population_%': [35, 82, 83, 88, 52, 91, 77, 43]
}

df = pd.DataFrame(data)

print("Dataset Preview:")
print(df)

plt.figure(figsize=(10, 6))
sns.heatmap(df.drop('Country', axis=1).corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap: Female Employment vs Socioeconomic Factors')
plt.show()

sns.pairplot(df.drop(columns=['Country']), corner=True)
plt.suptitle('Female Labor Participation vs Various Socioeconomic Factors', y=1.02)
plt.show()

plt.figure(figsize=(8, 6))
sns.regplot(data=df, x='GDP_per_Capita', y='Female_Labor_Participation_Rate')
plt.title('Female Labor Participation vs GDP per Capita')
plt.xlabel('GDP per Capita (USD)')
plt.ylabel('Female Labor Participation Rate (%)')
plt.grid(True)
plt.show()
```
